```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

try:
  result = requests.get ("https://www.cinemais.com.br/guaratingueta/programacao")
except:
  print('Não encontrou o site')
```

```python
if (result.status_code != 200):
  print('Erro ao carregar o site')
```

```python
pagina = BeautifulSoup(result.text)
```

```python
conteudo = pagina.select("div[class*='aba_0']")
```

**O select retorna uma lista de elementos. Então, vamos pegar somente o primeiro da lista porque só tem um container de programação da página.**

```python
grade = conteudo[0]
```

```python
filmes = grade.select("div[class='filme']")
```

```python
len(filmes)
```

```python
df = pd.DataFrame(columns=["titulo","sala","legenda","aspecto"])
for filme in filmes:
  titulo =filme.find("h5").get_text()
  #print(titulo)

  salas = filme.select("div.sala-filme div b")
  salaTexto = sala[0].get_text()
  #print(salaTexto)

  salaLegenda = filme.select("div.sala-filme > div")
  primeiroDiv = salaLegenda[0]
  filhosPrimeiroDiv = primeiroDiv.contents
  legenda = filhosPrimeiroDiv[2] #pega p terceiro elemento da lista
  #print(legenda)

  legendaSeparada = legenda.split('-')
  #print(legendaSeparada)

  legendadoDublado = legendaSeparada[1].strip()
  aspecto = legendaSeparada[2].strip()
  #print(legendadoDublado)
  #print(aspecto)

  #Adiciona um filme em cada linha do DataFrame
  df.loc[len(df)] = [ titulo, salaTexto, legendadoDublado, aspecto ]
```
